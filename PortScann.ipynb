```python
import socket



def scan_port(host, port):

    try:
        # Cria um objeto socket
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(0.1)  # Define um timeout para a conexão

        # Tenta se conectar ao host e porta especificados
        result = sock.connect_ex((host, port))

        # Verifica o resultado da conexão
        if result == 0:
            print(f"Porta {port} está aberta")
            return True  # Retorna True se encontrar uma porta aberta
        else:
            print(f"Porta {port} está fechada")
        sock.close()

    except Exception as e:
        print(f"Ocorreu um erro: {e}")



    return False  # Retorna False se não encontrar uma porta aberta



# Define o host que deseja escanear

host = "10.3.77.17"  # Pode ser substituído pelo IP do seu servidor



# Escaneia as portas de 1000 a 5000 | 3306 é cursed

for port in range(5000, 1000, -1):

    print(f"Verificando porta {port}")

    if scan_port(host, port):

        print(f"Porta aberta {host}, {port}")

        break  # Para o loop se encontrar uma porta aberta
```
